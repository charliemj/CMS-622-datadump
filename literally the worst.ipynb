```python
def get_tag_freq(csv_file):
    content = []
    with open(csv_file) as f:
        for line in f.readlines():
            content.append(line)
        f.close()
    tag_freq = {}
    for i in content[1:]:
        sp = i.split(',')
        num_notes = sp[2]
        tag_set = [sp[3][1:].lower()] + [i.lower() for i in sp[4:-1]] + [sp[-1][:-3].lower()]
        for t in tag_set:
            if t in tag_freq:
                tag_freq[t] += 1
            else:
                tag_freq[t] = 1
    print len(content)
    return tag_freq
```
